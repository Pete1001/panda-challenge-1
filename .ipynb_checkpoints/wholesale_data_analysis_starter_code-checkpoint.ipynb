## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['line_subtotal'] = df['unit_price'] * df['qty']
df['line_subtotal']
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda weight: 7 * weight if weight > 50 else 10 * weight)
df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['line_price_before_tax'] = df['line_subtotal'] + df['shipping_price']
df['line_price'] = df['line_price_before_tax'] * 1.0925
df[['line_subtotal', 'shipping_price', 'line_price']].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

#df['line_subtotal'] = df['unit_price'] * df['qty']
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda weight: 7 * weight if weight > 50 else 10 * weight)
df['line_price_before_tax'] = df['line_subtotal'] + df['shipping_price']
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
columns_to_display = [
    'first', 'last', 'job', 'phone', 'email', 'client_id', 'order_id', 'order_date', 'order_week', 'order_year',
    'unit_price', 'unit_cost', 'unit_weight', 'qty', 'line_number',
    'line_subtotal', 'total_weight', 'shipping_price', 'line_price', 'line_cost'
]
df[columns_to_display].head(3)  # Adjust .head() to display more rows if needed
```

```python
# Create a column for the profit of each line using line cost and line price
df['total_weight'] = df['unit_weight'] * df['qty']
df['line_price_before_tax'] = df['line_subtotal'] + df['shipping_price']
df['line_price'] = df['line_price_before_tax'] * 1.0925
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
df['line_profit'] = df['line_price'] - df['line_cost']
columns_to_display = [
    'first', 'last', 'job', 'phone', 'email', 'client_id', 'order_id', 'order_date', 'order_week', 'order_year',
    'unit_cost', 'unit_weight', 'qty', 'line_number',
    'line_subtotal', 'total_weight', 'shipping_price', 'line_price', 'line_cost', 'line_profit'
]
df[columns_to_display].head(3)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
summary_top_5 = df[df['client_id'].isin(top_clients_list)].groupby('client_id').agg(
    total_units=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('line_price', 'sum'),
    total_profit=('line_profit', 'sum')
).reset_index()

# Display the summary DataFrame
summary_top_5
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_columns = ['total_revenue', 'total_profit', 'total_shipping_price']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
    return value / 1_000_000

# Apply the currency_format_millions function to only the money columns. 
summary_top_5['total_revenue_million'] = summary_top_5['total_revenue'].apply(currency_format_millions)
summary_top_5['total_profit_million'] = summary_top_5['total_profit'].apply(currency_format_millions)
summary_top_5['total_shipping_million'] = summary_top_5['total_shipping_price'].apply(currency_format_millions)
summary_top_5['total_cost_million'] = summary_top_5['total_revenue_million'] - summary_top_5['total_profit_million']

# Rename the columns to reflect the change in the money format. 
summary_top_5 = summary_top_5.rename(columns={
    'total_revenue_million': 'Total Revenue (Million $)',
    'total_profit_million': 'Total Profit (Million $)',
    'total_shipping_million': 'Total Shipping (Million $)',
    'total_cost_million': 'Total Cost (Million $)',
    'total_units': 'Total Units Purchased'
})
summary_top_5[['client_id', 'Total Units Purchased', 'Total Shipping (Million $)', 'Total Revenue (Million $)', 'Total Cost (Million $)', 'Total Profit (Million $)']]

sorted_summary[['client_id', 'Total Units Purchased', 'Total Shipping (Million $)', 'Total Revenue (Million $)', 'Total Cost (Million $)', 'Total Profit (Million $)']]
```
